# DATA CLEANING AND MANIPULATION

```python
# filtered the Data set for Amazon
amazon = ['AMZN']
Data_sp_filtered = Data_sp[Data_sp['Symbol'].isin(amazon)]
```

```python
Data_sp_filtered['Symbol'].unique()
```

```python
Data_sp_filtered["Date"].unique()
```

```python
# coverting the Date column to datetime 
Data_sp_filtered['Date'] = pd.to_datetime(Data_sp_filtered['Date'])
```

```python
# filtering the data set by specific time frame
filtered_amazon = Data_sp_filtered[Data_sp_filtered['Date'].between('2020-01-01', '2024-12-31')]
```

# DATA VISUALIZATION

```python
print(filtered_amazon.shape)
```

```python
filtered_amazon = filtered_amazon.sort_values('Date')
```

```python
plt.plot(filtered_amazon['Date'], filtered_amazon['Close'], label = 'Amazon (AMZN)')
plt.xlabel('Date')
plt.ylabel('Close price')
plt.title('Amazon Stock Price Trend (2020-2024)')
plt.legend()
plt.show()
```

# CONCLUSION
The analysis of Amazon’s stock price trends from 2020 to 2024 reveals a significant degree of volatility. The stock price saw a sharp rise from 2020 to 2021, reaching a peak of around $180 before experiencing a downturn in 2022, where it dropped to nearly $100. Despite this decline, the stock began recovering in late 2023, and by 2024, Amazon’s stock price had climbed back to over $200. This reflects the resilience of Amazon’s stock, as it has consistently bounced back from declines over the years.
